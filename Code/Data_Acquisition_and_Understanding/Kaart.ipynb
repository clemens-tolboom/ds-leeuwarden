```python
# http://jupyter.org/widgets
# https://ipyleaflet.readthedocs.io/en/latest/api_reference/marker.html
# https://github.com/jupyter-widgets/ipyleaflet

from ipyleaflet import Map, CircleMarker, Marker, basemaps, basemap_to_tiles, Rectangle, Polygon

from pandas import DataFrame, read_csv
import pandas as pd
```

```python
root_dir = '../..'

sensors_file = root_dir + '/Sample_Data/For_Modeling/sensor_gps_view.csv'
sensors = pd.read_csv(sensors_file, delimiter=';', decimal=',')

center = (sensors['longitude'].mean(), sensors['latitude'].mean())
sensors
```

```python
import numpy as np

def arrow (p1, p2):
    y1, x1 = p1
    y2, x2 = p2

    # direction vector
    dx = x2 - x1
    dy = y2 - y1

    # normal vector
    nx = -dy
    ny = dx
    
    scale_y = .35
    dy = dy * scale_y
    ny = ny * scale_y
    
    return [
        (y1, x1),
        (y2, x2),
        (y2 + ny * 0.05,                         x2 + nx * 0.05),
        (y2 + ny * 0.05 + (ny + dy) * 0.01,      x2 + nx * 0.05 - (nx + dx) * 0.01),
#        (y1 + ny,                  x1 + nx),
#        (p2[0] + ny * 0.05 - (ny + dy) / 16, p2[1] + nx * 0.05 - (nx + dx) / 16)
    ]

```

```python
m = Map(
    center=center,
    zoom=16
)

for index, row in sensors.iterrows():
    if index> 0:
        # print( index, row)
        form = (row.longitude, row.latitude)
        marker = Marker(location=form, draggable=False)
        m.add_layer(marker);
        for j, r in sensors.iterrows():
            if j>0 and j != index:
                to = (r.longitude, r.latitude)
                arrow(form, to)
                p = Polygon(
                    locations= arrow(form, to),
                    color="red",
                    fill_color="orange"
        
                )
                m.add_layer(p)
            break


m
```

```python
for i in range(20):
    a = arrow( center, (center[0]+0.002 * i/10, center[1]+0.003))

    p = Polygon(
        locations= a,
        color="red",
        fill_color="orange"
    )

    m.add_layer(p)
```
